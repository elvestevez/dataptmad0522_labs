```python
def rename_class(name):
    new_name = name
    if name == 'Iris-setossa':
        new_name = 'Iris-setosa'
    elif name == 'versicolor':
        new_name = 'Iris-versicolor'
    return new_name
```

```python
# correct value of feature class
df_iris['class'] = df_iris.apply(lambda x: rename_class(x['class']), axis=1)
df_iris.groupby('class').size()
```

```python
# option 1: dropna
print(len(df_iris))
df_iris_new = df_iris.dropna()
print(len(df_iris_new))
df_iris_new.isnull().sum()
```

```python
# option 2: assign value mean for same class
print(f"Total mean: {df_iris['petal_width_cm'].mean()}")
print(f"Iris-setosa mean: {df_iris.loc[df_iris['class'] == 'Iris-setosa']['petal_width_cm'].mean()}")
```

```python
def set_n_class(name):
    if name == 'Iris-setosa':
        return 1
    elif name == 'Iris-versicolor':
        return 2
    elif name == 'Iris-virginica':
        return 3
```

```python
df_iris_size['n_class'] = df_iris_size.apply(lambda x: set_n_class(x['class']), axis=1)
```
