#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, axs = plt.subplots(ncols=1,
                        nrows=2,
                        figsize=(12,10))

axs[0].plot(x, y)
axs[0].plot(x, z)
axs[0].set(title='Challenge 1 - normal scale')

axs[1].plot(x, y)
axs[1].plot(x, z)
axs[1].set(title='Challenge 1 - log scale')
axs[1].set_xscale('log')
axs[1].set_yscale('log')

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, axs = plt.subplots(ncols=1,
                        nrows=2,
                        figsize=(12,10))

axs[0].plot(x, y, label='pow')
axs[0].plot(x, z, label='exp')
axs[0].legend(loc='upper left')
axs[0].set(title='Challenge 1 - normal scale')

axs[1].plot(x, y, label='pow', color='green')
axs[1].plot(x, z, label='exp', color='red')
axs[1].legend(loc='upper left')
axs[1].set(title='Challenge 1 - log scale')
axs[1].set_xscale('log')
axs[1].set_yscale('log')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

```python
fitbit.dtypes
fitbit['Work_or_Weekend'].unique()
display(fitbit[fitbit['Work_or_Weekend'] == 0].head(3)) # weekend
display(fitbit[fitbit['Work_or_Weekend'] == 1].head(3)) # work
```

```python
fitbit_month = fitbit.groupby(['Months','Months_encoded'])['Steps'].mean().reset_index().sort_values('Months_encoded')
fitbit_month
```

```python
def set_work_weekend(w):
    if w == 0:
        return 'weekend'
    elif w == 1:
        return 'work'

fitbit_week = fitbit.groupby(['Work_or_Weekend'])['Steps'].mean().reset_index()
fitbit_week['work_weekend'] = fitbit_week.apply(lambda d: set_work_weekend(d['Work_or_Weekend']), axis = 1)
fitbit_week
```

```python
fig_f, axs_f = plt.subplots(ncols=1,
                            nrows=2,
                            figsize=(12,8))

fig_f.suptitle('Activity', fontweight='bold') 

plt.subplots_adjust(hspace=0.5)

# MEDIAN STEPS BY MONTH_ENCODED
# your code here
plt.sca(axs_f[0])
plt.bar(fitbit_month['Months'], fitbit_month['Steps'])
plt.title('Median steps')
plt.xticks(rotation='vertical')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
plt.sca(axs_f[1])
plt.title('Median steps')
plt.barh(fitbit_week['work_weekend'], fitbit_week['Steps'])

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit.head(2)
```

```python
# your code here
fig_f, axs_f = plt.subplots(ncols=1,
                            nrows=3,
                            figsize=(12,10))

fig_f.suptitle('Activity vs Steps', fontweight='bold') 

plt.subplots_adjust(hspace=0.5)

for i in range(3):
    plt.sca(axs_f[i])
    if i == 0:
        # Minutes Sedentary vs Steps
        x = fitbit['Minutes Sedentary']
        y = fitbit['Steps']
        color = 'blue'
        size = 8
        mark = 'o'
        title = 'Sedentary'
    elif i == 1:
        # Minutes Sedentary vs Steps
        x = fitbit['Minutes Lightly Active']
        y = fitbit['Steps']
        color = 'green'
        size = 25
        mark = 's'
        title = 'Lightly Active'
    elif i == 2:
        # Minutes Lightly Active vs Steps
        x = fitbit['Minutes Very Active']
        y = fitbit['Steps']
        color = 'orange'
        size = 35
        mark = '^'
        title = 'Very Active'
    
    plt.scatter(x, y, c=color, s=size, marker=mark)
    plt.title(title)
    plt.xlabel('Minutes')
    plt.ylabel('Steps')

plt.show()
```
